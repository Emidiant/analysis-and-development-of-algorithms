# Алгоритмы на графах. Алгоритмы поиска пути на взвешенных графах

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import matplotlib.colors as mcolors
import time
```

## Part 1

```python
n = 100
m = 500
matrix = np.zeros((n, n))
count = 0
while count != m:
    i = random.randint(0, n - 1)
    j = random.randint(0, n - 1)
    if matrix[i][j] == 0 and i != j:
        weight = random.randint(1, 100)
        matrix[i][j] = weight
        matrix[j][i] = weight
        count += 1

g = nx.from_numpy_matrix(matrix)
```

```python
def generate_adjlist_with_all_edges(G):
    for s, nbrs in G.adjacency():
        line = str(s) + ": "
        list_edges = []
        for t, data in nbrs.items():
                line += str(t) + " "
                list_edges.append(t)
        yield list_edges
        
adjlist = []
for line in generate_adjlist_with_all_edges(g):
    adjlist.append(line)
    
adjlist_dict = {}
for i in range(len(adjlist)):
    adjlist_dict[i] = adjlist[i]
```

```python
 def show_large_graph(graph):
    plt.figure(num=None, figsize=(20, 20), dpi=80)
    plt.axis('off')
    fig = plt.figure(1)
    pos = nx.spring_layout(graph)
    nx.draw_networkx_nodes(graph, pos)
    nx.draw_networkx_edges(graph, pos)
    nx.draw_networkx_labels(graph, pos)
    
    nx.draw_networkx_edge_labels(g, pos)

    plt.show()

show_large_graph(g)
```

```python
def time_analize(node_start, algorithm):
    time_ex = []
    for i in range(10):
        start_time = time.time()
        algorithm
        path = algorithm(g, node_start)
        time_ex.append(time.time() - start_time)
    print(round(sum(time_ex) / len(time_ex), 5))
    return path
```

### Dijkstra's algorithm

find shortest paths between a random starting vertex and other vertices

```python
node_start = random.randint(0, n - 1)
```

```python
time_analize(node_start, nx.single_source_dijkstra_path)
```

### Bellman-Ford algorithm

find shortest paths between a random starting vertex and other vertices

```python
time_analize(node_start, nx.single_source_bellman_ford_path)
```

## Part 2

Generate a 10x20 cell grid with 40 obstacle cells. Choose two random non- obstacle cells and find a shortest path between them using A* algorithm. Repeat the experiment 5 times with different random pair of cells. Analyse the results obtained.

```python
data = np.ones((10, 20), dtype=int)
list_obstacles = []
while data.sum() != 160:
    i = random.randint(0, 9)
    j = random.randint(0, 19)
    if data[i][j] != 0:
        list_obstacles.append((i, j))
        data[i][j] = 0
```

```python
data
```

```python
G = nx.grid_2d_graph(10,20)
G.remove_nodes_from(list_obstacles)
```

```python
def get_nodes():
    while True:
        i = random.randint(0, 9)
        j = random.randint(0, 19)
        start_node = (i, j)
        i = random.randint(0, 9)
        j = random.randint(0, 19)
        end_node = (i, j)
        if start_node in G and end_node in G:
            print(start_node, end_node)
            break
    return start_node, end_node

def draw_astar_path(path_astar):
    colors = list(mcolors.TABLEAU_COLORS)
    color_map = []

    for node in G:
        if node in path_astar:
            color_map.append(colors[2])
        else:
            color_map.append(colors[0])

    plt.figure(figsize=(16,8))
    pos = {(x,y):(y,-x) for x,y in G.nodes()}
    nx.draw(G, pos=pos, 
            node_color=color_map, 
            with_labels=True,
            node_size=1400)
    plt.show()
```

```python
for i in range(5):
    start_node, end_node = get_nodes()
    try:
        path_astar = nx.astar_path(G, start_node, end_node)
        draw_astar_path(path_astar)
    except nx.NetworkXNoPath:
        print(f"Node {end_node} not reachable from {start_node}")
```
